```python
import pandas as pd
```

```python
employees = pd.read_csv("data/employees.csv")
employees.head(2)
```

```python
employees["insurance"] == True
yes_insurance = employees[employees["insurance"] == True]
no_insurance = employees[employees["insurance"] == False]
no_insurance
```

```python
# context manager
with pd.ExcelWriter("insurance-results.xlsx") as excel_file:
    yes_insurance.to_excel(excel_file, sheet_name="insurance-yes", index=False)
    no_insurance.to_excel(excel_file, sheet_name="insurance-no", index=False, columns=["gender", "insurance", "salary"])
```
