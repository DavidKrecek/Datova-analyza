Pokračování - filtrování dat pomocí a zároveň

```python
is_female = employees.gender == "Female"
is_in_support = employees.department == "Support"
salary_over_500k = employees.salary >= 500000
have_insurance = employees.insurance == True
```

```python
is_female & is_in_support
employees[is_female & is_in_support]
```

```python
employees[is_female & have_insurance]
```

```python
employees[is_female & is_in_support & salary_over_500k]
```

úkol - všichni muži, kteří mají větší mzdu než 600k a jsou to prodejci (sales)

```python
is_male = employees.gender == "Male"
salary_over_600k = employees.salary >= 600000
is_in_sales = employees.department == "Sales"
```

```python
employees[is_male & salary_over_600k & is_in_sales]
```

```python
# znova jsem si vytáhnul připravené sloupce

is_female = employees.gender == 'Female'
is_in_support = employees.department == 'Support'
salary_over_500k = employees.salary >= 500000
have_insurance = employees.insurance == True
is_male = employees.gender == 'Male'
salary_over_600k = employees.salary >= 600000
is_in_sales = employees.department == 'Sales'
```

```python
employees[is_in_support | salary_over_600k]
```
