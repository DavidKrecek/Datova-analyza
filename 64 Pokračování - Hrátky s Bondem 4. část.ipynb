```python
import pandas as pd
```

```python
james_bond_df = pd.read_csv("data/jamesbond.csv")
james_bond_df.head()
```

```python
james_bond_df.info()
```

```python
james_bond_df.describe()
```

```python
1965 // 10 * 10
1982 // 10 * 10
1995 // 10 * 10
```

```python
james_bond_df["Decade"] = (james_bond_df["Year"] // 10) * 10
james_bond_df.head(3)
```

```python
movies_per_decade = james_bond_df["Decade"].value_counts().sort_index()
movies_per_decade
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 3))
movies_per_decade.plot(kind="bar", color="blue")
plt.title("Number of Bond movies released per decade")
plt.xlabel("Decade")
plt.ylabel("Number of movies")
plt.show()          # dle mě toto není třeba, ani pro CS code
```

```python
avg_rating_by_actor = james_bond_df.groupby("Bond")["Avg_User_IMDB"].mean().sort_values(ascending=False)
avg_rating_by_actor
```

```python
avg_budget_by_actor = james_bond_df.groupby("Bond")["Budget"].max().sort_values(ascending=False)
avg_budget_by_actor
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5, 3))
avg_rating_by_actor.plot(kind="barh", color="green")
plt.title("Average IMDB rating by Bond actor")
plt.xlabel("Average IMDB rating")
plt.xlabel("Bond actor")
plt.show()
```

Hrátky s Bondem 3. část

```python
from collections import Counter
all_locations = ", ".join(james_bond_df["Shooting_Loc"]).split(", ")
location_counts = Counter(all_locations)
location_counts

top_location = pd.Series(location_counts).sort_values(ascending=False).head(10)
top_location
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 3))
top_location.plot(kind='bar', color="green")
plt.title('Top 10 Bond movie shooting location')
plt.xlabel('Location')
plt.ylabel('Number of movies')
plt.show()
```
