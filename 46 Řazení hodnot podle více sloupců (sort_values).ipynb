```python
import pandas as pd
```

```python
students = pd.read_csv("data/students.csv").dropna(how="all")
```

```python
students.info()
```

```python
students
```

```python
students.favourite_color = students.favourite_color.fillna("unknown")
students.weight = students.weight.fillna(0)
students.height = students.height.fillna(0)
students.info()
```

```python
students.sort_values(by="favourite_color").head(10)
students.sort_values("favourite_color").head(10)
```

```python
students.sort_values(["favourite_color", "weight"]).head(10)
```

```python
students.sort_values(["favourite_color", "weight"], ascending=False).head(10)
```

```python
students.sort_values(["favourite_color", "weight"], ascending=[True, False]).head(10)
```

```python
sorted_students = students.sort_values(["favourite_color", "weight"], ascending=[True, False])
sorted_students
```
