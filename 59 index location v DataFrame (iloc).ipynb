iloc

```python
import pandas as pd
```

```python
bond = pd.read_csv("data/jamesbond.csv")
bond.head(3)
```

index location - iloc

```python
bond.iloc[2]        # Series
bond.iloc[[3, 3]]   # DataFrame
```

```python
bond.iloc[4:8]  # 4-7, 8 není součástí
```

```python
bond.iloc[0:4]
bond.iloc[:4]       # stejný zápis jako výše
```

```python
bond.iloc[20:]  # od 20 do konce
```

```python
bond.iloc[-1]   # vrací poslední film
```

```python
bond.iloc[-2]   # přeposlední film
```

```python
bond.iloc[0:10:2]
bond.iloc[:10:2]        # stejný výsledek jako výše
```
