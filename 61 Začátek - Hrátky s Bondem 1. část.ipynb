```python
james_bond_df["Decade"] = (james_bond_df["Year"] // 10) * 10
james_bond_df.head(3)
```

```python
movies_per_decade = james_bond_df["Decade"].value_counts().sort_index()
movies_per_decade
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 3))
movies_per_decade.plot(kind="bar", color="blue")
plt.title("Number of Bond movies released per decade")
plt.xlabel("Decade")
plt.ylabel("Number of movies")
plt.show()          # dle mÄ› toto nenÃ­ tÅ™eba, ani pro CS code
```
