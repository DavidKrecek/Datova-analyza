```python
# iloc - vybírá na základě pozice
# toto skončí chybou: bond.iloc['From Russia with Love']
# nezahrnuje konečný bod rozsahu (je exkluzivní na konci): bond.iloc[2:10]

# loc - vybírá na základě označení (např. názvu filmu)
# toto skončí chybou: bond.loc[0]
# zahrnuje oba konce rozsahu (je inkluzivní)
```

```python
import pandas as pd
```

```python
bond = pd.read_csv("data/jamesbond.csv")
bond.head(3)
```

```python
bond.iloc[0]
bond.iloc[[0, 1]]
```

```python
bond = pd.read_csv("data/jamesbond.csv", index_col="Movie")
bond.iloc[0]    
bond.loc["Dr. No"]  # Series
bond.loc["Goldfinger"]  # Series
# bond.loc["Martinis man"]    # když neexistuje, tak to vyhodí chybu
```

```python
bond.loc[["Dr. No", "Spectre"]]     #DataFrame
# bond.loc[["Dr. No", "Spectre", "Pokemon"]]  # vyhodí chybu, i když neex. jeden
```

```python
bond.loc["Casino Royale": "Spectre"] # najde hodnoty i mezi vč. posledního
```

```python
bond.loc["Spectre": "Casino Royale"] # vrátí prádzný DataFrame = špatné pořadá
```

```python
bond.loc["Casino Royale":]  # od casino do konce
bond.loc[:"Goldfinger"]     # od začátko po Goldfinger
```
